```python
import numpy as np
import matplotlib.pyplot as plt
```

# I use gamma distribultion with shape parameters k=2.0, Î¸=2.0

```python
gamma = np.random.gamma(shape=2.0, scale=2.0, size=1000)

plt.hist(gamma, bins=100)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Gamma Distributed Random Numbers')
plt.savefig("session10.pdf")
plt.show()
```
